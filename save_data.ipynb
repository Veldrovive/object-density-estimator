```python
from img_preprocessing import get_full_frame
import matplotlib.pyplot as plt
import numpy as np
```

```python
blur = True
scale_down = 4
blur_rad = 2
```

```python
frame = get_full_frame(0, blur=blur, scale_down=scale_down, blur_rad=blur_rad)
```

```python
plt.imshow(frame["image"])
```

```python
loc_img = frame["loc_img"]
loc_img = loc_img.reshape(loc_img.shape[0], loc_img.shape[1])
print(loc_img.shape)
print(np.sum(loc_img))
plt.imshow(loc_img)
```

```python
img_arr = np.zeros((1000, 480, 640, 3))
den_arr = np.zeros((1000, 480//scale_down, 640//scale_down, 1))
for i in range(1000):
    print("Iteration: {}".format(i))
    frame = get_full_frame(i, blur=blur, scale_down=scale_down, blur_rad=blur_rad)
    img_arr[i] = frame["image"]
    den_arr[i] = frame["loc_img"].reshape((frame["loc_img"].shape[0], frame["loc_img"].shape[1], 1))
```

```python
#np.save("./data/mall_processed/train-imgs", img_arr[:750])
np.save("./data/mall_processed/train-densities-small-blur-regular", den_arr[:750])
#np.save("./data/mall_processed/val-imgs", img_arr[750:])
np.save("./data/mall_processed/val-densities-small-blur-regular", den_arr[750:])
```
